## Model development

- Now I can use the preprocessed dataset to train machine learning models to predict the house price given feature values.
- Actually, we have a little more preprocessing to do before the dataset gets fed into our models.

1. Postcode, Suburb, CouncilArea columns are dropped.
    - Suburb is a categorical variable with too many levels. This will make the training of models extremely slow.
    - Postcode and CouncilArea are also removed for similar reason.
2. Continous-valued features like Distance, Propertycount, DistTrain are normalized using MinMaxScaler().
3. Categorical variables are encoded using One-Hot-encoding.
4. Regionname is removed as it seems not too useful given that we have the Suburb feature.

```python
houses.drop(['Suburb', 'Postcode', 'CouncilArea'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

houses[['Distance', 'Propertycount', 'DistTrain']] = scaler.fit_transform(houses[['Distance', 'Propertycount', 'DistTrain']])
```

```python
houses = pd.get_dummies(houses)

X = houses.drop(['Price'], axis=1)
y = houses['Price']
```

```python
#pickle.dump(houses, open('data/houses_modelDev.pkl', 'wb'))
#pickle.dump(X, open('data/X_modelDev.pkl', 'wb'))
#pickle.dump(y, open('data/y_modelDev.pkl', 'wb'))
houses = pickle.load(open('data/houses_modelDev.pkl', 'rb'))
X = pickle.load(open('data/X_modelDev.pkl', 'rb'))
y = pickle.load(open('data/y_modelDev.pkl', 'rb'))
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

- 80-20 split for train and test sets.
- We will use cross validation so no need to explicitly have a validation set.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

### Baseline

- Let's use a simple linear regression as a baseline.

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

y_pred = lin_reg.predict(X_test)
```

```python
from sklearn import metrics

r2 = metrics.r2_score(y_test, y_pred)
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)

scores = pd.DataFrame({'Baseline (regression)' : [r2, mae, mse]}, index=['R2', 'MAE', 'MSE'])
scores
```

- An MAE of 217667 says that our linear regression model is on average $\$$217667 wrong on the test set. 

```python
diff = y_test - y_pred

plt.scatter(y_test, diff)
plt.ylabel('y_test - y_pred')
plt.xlabel('y_test')
plt.show()
```

- Ideally the y-values should be close to 0.
- We see that for more expensive houses (further to the right on the x-axis), we tend to have large difference, meaning we are overestimating the prices.
